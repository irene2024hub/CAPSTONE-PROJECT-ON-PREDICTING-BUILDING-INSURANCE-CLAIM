# DATASET DESCRIPTION

.The dataset contains historical records of insured buildings, including their physical attributes, location,
and claim status. Each row represents a unique insurance policy for a building observed over a specific period. 
The dataset includes both numerical and categorical features, which require preprocessing and transformation for modeling.
        
Feature Descriptions:
- Customer Id: Unique identifier for the policyholder.
- YearOfObservation: The year in which the insurance policy was observed.
- Insured_Period: Duration of the insurance policy. For example, 1 indicates a full year, while 0.5 indicates a 6-month policy.
- Residential: Indicates whether the building is residential (Yes) or not (No).
- Building_Painted: Indicates whether the building is painted (Yes) or not (No).
- Building_Fenced: Indicates whether the building is fenced (Yes) or not (No).
- Garden: Indicates whether the building has a garden (Yes) or not (No).
- Settlement: Type of area where the building is located — either Urban or Rural.
- Building Dimension: Size of the building in square meters (m²).
- Building_Type: Categorical variable indicating the type of building (Type 1, Type 2, Type 3, or Type 4).
- Date_of_Occupancy: The date the building was first occupied.
- NumberOfWindows: Total number of windows in the building.
- Geo Code: Geographical code representing the location of the building.

#### Categorical Variables are  non-numeric or discrete features that represent categories:

# Encode categorical variables using **One-hot encoding**

is a method used to convert categorical data into a 
 numerical format that machine learning models can understand.

**The following features will be encoded;**

- Residential (e.g., Yes/No or 0/1)
- Building_Painted (e.g., Yes/No)
- Building_Fenced (e.g., Yes/No)
- Garden (e.g., Yes/No)
- Settlement (e.g., Urban/Rural)
- Building_Type (e.g., Type1, Type2, etc.)
- Geo_Code (e.g., region codes)



    Why We Encoded 'V' and 'N' as 1 and 0
In our dataset, some features like Residential, Building_Painted, Building_Fenced, and Garden use the letters 'V' and 'N' to represent Yes and No. Since machine learning models work best with numbers (not text), we converted these values into numeric form:
- 'V' (Yes) → 1
- 'N' (No) → 0
This makes the data easier for the model to understand and helps it learn patterns more effectively. It's a simple but important step in preparing categorical data for analysis or prediction.



````python
# Label encode binary categorical features 
df = pd.read_csv("Train_data.csv")
binary_map = {'V': 1, 'N': 0}

df['Residential'] = df['Residential'].map(binary_map)
df['Building_Painted'] = df['Building_Painted'].map(binary_map)
df['Building_Fenced'] = df['Building_Fenced'].map(binary_map)
df['Garden'] = df['Garden'].map(binary_map)
````

````python
# If not already done:
df['Settlement'] = df['Settlement'].map({'U': 'Urban', 'R': 'Rural'})
df = pd.get_dummies(df, columns=['Settlement'], prefix='Settlement', drop_first=True)

````

#### WHY DROPPED GEO_CODE FROM THE MODEL

* GEO_CODE* was removed from the dataset because it had **over 1,300 unique values**, making it a high-cardinality categorical feature. Including it in its raw form would have required one-hot encoding, which creates a separate binary column for each unique value. This process alone caused the number of columns in our dataset to **balloon from 14 to 1,321**, introducing over 1,300 new columns just for Geo_Code.
This kind of dimensionality explosion can lead to several issues:
- Overfitting: The model may memorize noise instead of learning general patterns.
- Increased computational cost: More features mean longer training times and higher memory usage.
- Sparse data: Most of the new columns contain mostly zeros, which dilutes meaningful signals.
Since Geo_Code is essentially a location identifier with no inherent numerical meaning, and we lacked additional geographic context (like regional groupings or coordinates), we determined that its inclusion would add complexity without guaranteed predictive value. Therefore, we chose to drop it to keep the model simpler, faster, and more generalizable.


````python

# Step 1: Identify all one-hot encoded Geo_Code columns
geo_cols = [col for col in df.columns if col.startswith('Geo_Code_')]

# Step 2: Drop those columns from the DataFrame
df.drop(columns=geo_cols, inplace=True)

# Step 3: Confirm the new shape of your dataset
print("Corrected shape:", df.shape)
````
